### Scoring guide:

Your total score is the sum of your scores for each round.
The score for a single round is the score for the shape you selected (1 for Rock, 2 for Paper, and 3 for Scissors)
plus the score for the outcome of the round (0 if you lost, 3 if the round was a draw, and 6 if you won).

```python
WINNING_MAP = {
    'A': 'Y',  # Rock < Paper
    'B': 'Z',  # Paper < Scissors
    'C': 'X',  # Scissors < Rock
}
SHAPE_SCORING = {
    'X': 1,  # Rock
    'Y': 2,  # Paper
    'Z': 3,  # Scissors
}
LOSE, DRAW, WON = (0, 3, 6)
OUTCOME_SCORING = {
    'A Y': WON,
    'A X': DRAW,
    'B Z': WON,
    'B Y': DRAW,
    'C X': WON,
    'C Z': DRAW
}

scores = 0
for r in rounds:
    scores += OUTCOME_SCORING.get(r, 0)
    scores += SHAPE_SCORING[r[2]]

scores
```

```python
ROCK, PAPER, SCISSORS = (1, 2, 3)
UPDATED_SCORING = {
    # Rock
    'A': {
        'X': (LOSE, SCISSORS),
        'Y': (DRAW, ROCK),
        'Z': (WON, PAPER)
    },
    # Paper
    'B': {
        'X': (LOSE, ROCK),
        'Y': (DRAW, PAPER),
        'Z': (WON, SCISSORS)
    },
    # Scissors
    'C': {
        'X': (LOSE, PAPER),
        'Y': (DRAW, SCISSORS),
        'Z': (WON, ROCK)
    }
}
scores = 0
for r in rounds:
    opponent, player = r.split(' ')
    scores += sum(UPDATED_SCORING[opponent][player])
scores
```
